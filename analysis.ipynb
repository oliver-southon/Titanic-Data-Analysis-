```python
import pandas as pd
import numpy as np
```

```python
# Load in data

data = pd.read_csv("Titanic-1.csv", index_col=0)
data.dropna()
data.head()
```

```python
from sklearn.model_selection import train_test_split # splitting data
from sklearn.tree import DecisionTreeClassifier # decision tree
from sklearn import metrics # for accuracy
```

```python
# Split data

feature_cols = ["PClass", "Age", "Sex"] # select features
X = data[feature_cols].values # features
y = data.Survived # target

# Change PClass from str to loat
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
for i in range(len(X[0])):
    X[:, i] = Le.fit_transform(X[:, i]) # convert row to nums
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Build model

clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
# Measure predictions

print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
# Measure predictions

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
# Tree Visualisation


from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image
import pydotplus

dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, filled=True, rounded=True, special_characters=True,
               feature_names = feature_cols, class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('titanic_survival.png')
Image(graph.create_png())
```
