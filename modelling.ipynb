```python
import pandas as pd
from matplotlib import pyplot as plt
from tabulate import tabulate
```

```python
df = pd.read_csv('Titanic-1.csv')
df.head()
```

```python
# ==HYPOTHESIS== 
# 
# Being a female on the Titanic would give you a higher chance of survival than being under the age of 20
```

```python
# >1. AGES OF ALL TITANIC PASSENGERS

survived = df.loc[df['Survived'] == 1]['Age']
not_survived = df.loc[df['Survived'] == 0]['Age']

fig, ax = plt.subplots(1, 1, figsize=(13,8))

ax.grid()
ax.set_axisbelow(True)
ax.hist([survived, not_survived], bins=range(0,75,5), label=["Survived", "Didn't Survive"], edgecolor="black", color=['forestgreen','cornflowerblue'])
ax.legend(loc="upper right")
plt.xlabel("Age", fontsize=15)
plt.ylabel("Frequency", fontsize=12)
plt.title("Ages of Titanic Passengers", fontsize=24)

table = [
    ["Min", survived.min(), not_survived.min()],
    ["Max", survived.max(), not_survived.max()],
    ["Median", survived.median(), not_survived.median()],
    ["Mean", survived.mean(), not_survived.mean()],
    ["Q1", survived.quantile(0.25), not_survived.quantile(0.25)],
    ["Q3", survived.quantile(0.75), not_survived.quantile(0.75)],
    ["IQR", survived.quantile(0.75)-survived.quantile(0.25), not_survived.quantile(0.75)-not_survived.quantile(0.25)]
]

print(tabulate(table, headers=[" ", "survived", "Not survived"], tablefmt='orgtbl'))
```

```python
# reset vars
survived = df.loc[df['Survived'] == 1]
not_survived = df.loc[df['Survived'] == 0]

# make new vars
m_survived = survived.loc[survived['Sex'] == 'male']['Age']
m_not_survived = not_survived.loc[not_survived['Sex'] == 'male']['Age']

# plot
fig, ax = plt.subplots(1, 1, figsize=(13,8))

ax.grid()
ax.set_axisbelow(True)
ax.hist([m_survived, m_not_survived], bins=range(0,75,5), label=["Survived", "Didn't Survive"], edgecolor="black", color=['forestgreen','cornflowerblue'])
ax.legend(loc="upper right")
plt.xlabel("Age", fontsize=15)
plt.ylabel("Frequency", fontsize=12)
fig.suptitle("Age of Male Titanic Passengers", fontsize=24)

table = [
    ["Min", m_survived.min(), m_not_survived.min()],
    ["Max", m_survived.max(), m_not_survived.max()],
    ["Median", m_survived.median(), m_not_survived.median()],
    ["Mean", m_survived.mean(), m_not_survived.mean()],
]

print(tabulate(table, headers=[" ", "survived", "Not survived"], tablefmt='orgtbl'))
```

```python
# reset vars
survived = df.loc[df['Survived'] == 1]
not_survived = df.loc[df['Survived'] == 0]

f_survived = survived.loc[survived['Sex'] == 'female']['Age']
f_not_survived = not_survived.loc[not_survived['Sex'] == 'female']['Age']

fig, ax = plt.subplots(1, 1, figsize=(13,8))

ax.grid()
ax.set_axisbelow(True)
ax.hist([f_survived, f_not_survived], bins=range(0,75,5), label=["Survived", "Didn't Survive"], edgecolor="black", color=['forestgreen','cornflowerblue'])
ax.legend(loc="upper right")
plt.xlabel("Age", fontsize=15)
plt.ylabel("Frequency", fontsize=12)
fig.suptitle("Age of Female Titanic Passengers", fontsize=24)

table = [
    ["Min", f_survived.min(), f_not_survived.min()],
    ["Max", f_survived.max(), f_not_survived.max()],
    ["Median", f_survived.median(), f_not_survived.median()],
    ["Mean", f_survived.mean(), f_not_survived.mean()],
]

print(tabulate(table, headers=[" ", "survived", "Not survived"], tablefmt='orgtbl'))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(13,8))

ax.grid()
ax.set_axisbelow(True)

f_tot_survived = len(f_survived)
f_tot_not_survived = len(f_not_survived)

x = ["Female Survivors", "Female Non-Survivors"]
y = [f_tot_survived, f_tot_not_survived]
ax.bar(x, y, edgecolor="black", color=['darkviolet','mediumpurple'])
plt.xlabel("Passengers", fontsize=15)
plt.ylabel("Count", fontsize=12)
fig.suptitle("Female Passengers That Survived And Did Not Survive", fontsize=24)

table = [
    ["Count", f_tot_survived, f_tot_not_survived]
]

print(tabulate(table, headers=[" ", "survived", "Not survived"], tablefmt='orgtbl'))
print("\nSurvival Percentage: "  + str(round(f_tot_survived/(f_tot_survived+f_tot_not_survived),4)))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(13,8))

ax.grid()
ax.set_axisbelow(True)

m_tot_survived = len(m_survived)
m_tot_not_survived = len(m_not_survived)

x = ["Male Survivors", "Male non-Survivors"]
y = [m_tot_survived, m_tot_not_survived]
ax.bar(x, y, edgecolor="black", color=['dodgerblue', 'lightskyblue'])
ax.set_xlabel("Passengers", fontsize=15)
ax.set_ylabel("Count", fontsize=12)
ax.set_title("Male Passengers That Survived And Did Not Survive", fontsize=24)

table = [
    ["Count", m_tot_survived, m_tot_not_survived]
]

print(tabulate(table, headers=[" ", "survived", "Not survived"], tablefmt='orgtbl'))
print("\nSurvival Percentage: "  + str(round(m_tot_survived/(m_tot_survived+m_tot_not_survived),4)))
```

```python
labels = ["Survivors", "Non-Survivors"]
men = [m_tot_survived, m_tot_not_survived]
women = [f_tot_survived, f_tot_not_survived]

width=0.7

fig, ax = plt.subplots(1, 1, figsize=(8,8))

ax.bar(labels, men, width, label='Men', color=['dodgerblue','lightskyblue'])
ax.bar(labels, women, width, bottom=men, label='Women', color=['darkviolet','mediumpurple'])
ax.set_ylabel('Count', fontsize=12)
ax.set_title('Survivors by Sex', fontsize=24)
ax.legend(fontsize=15)
print(m_tot_survived)
print(f_tot_survived)
```

```python
under_twenty = df.loc[df['Age'] < 20]
under_twenty_survived = under_twenty.loc[under_twenty['Survived'] == 1]
under_twenty_not_survived = under_twenty.loc[under_twenty['Survived'] == 0]

print("Chance of survival: " + str(round(len(under_twenty_survived)/len(under_twenty),4)))

labels = [''] 
  
mine = [len(under_twenty_survived)] 
others = [len(under_twenty_not_survived)] 

  
fig, ax = plt.subplots(1,1,figsize=(10,1)) 
  
ax.barh(labels, mine, label ='Survived', color='darkseagreen') 
  
ax.barh(labels, others,  left = mine, label ='Did not survive', color='palegreen') 
  
ax.set_xlabel('Passengers', fontsize=12) 
ax.legend() 
  
ax.set_title('Titanic Passengers Under 20', fontsize=24) 
  


```

```python
### IF HYPOTHESIS IS CORRECT, THEN...
#    P(Survive | Female) > P(Survive | less than 20)
#
# P(Surivve | Female)
pFemale = round(f_tot_survived/(f_tot_survived+f_tot_not_survived),4) 
print("If you are a female, your chance of survival is: " + str(pFemale))

# P(Surivve | Less than 20)
pAge = round(len(under_twenty_survived)/len(under_twenty),4)
print("If you are under the age of 20, your chance of survival is: " + str(pAge))

if pFemale > pAge:
    print("\nTherefore, the hypothesis is correct and being female gives you a higher chance of survival than being under the age of 20.")
else: 
    print("\nThe hypothesis has been disproved. You are more likely to survive being under 20 than being female.")
```
